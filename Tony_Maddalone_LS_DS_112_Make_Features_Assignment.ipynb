```python
# Lets do some feature engineering!
# We are going to reduce our dataset to 4 columns
# and make a smaller dataset

small_df = df[['TotalBsmtSF', '1stFlrSF', '2ndFlrSF', 'SalePrice']].copy() #copys orig dataset?

small_df
```

```python
# now let's address the extra NaN's in each column by
# skipping the footer as well

df = pd.read_csv('LoanStats_2018Q4.csv', header=1, skipfooter=2, engine='python')
print(df.shape)
df.head()
```

```python
# Let's sort the isnull values

df.isnull().sum().sort_values()
```

```python
# Now let's drop the 100% blank columns from above (128412's)

df = df.drop(['url', 'member_id', 'desc', 'id'], axis=1)

```

```python
df.head()
```

```python
# Let's clean the int_rate column

# Look at the first 10 values in int_rate column

df['int_rate'].head(10)
```

```python
# there are some problems with this column
# 1. string that should be numeric
# 2. percent % sign included in number
# 3. leading space at beginning of string

# 1. string that should be numeric

int_rate = ' 14.17'
```

```python
# Deal with leading space with .strip

int_rate.strip()
```

```python
# to remove both leading space and % sign

type(int_rate.strip().strip('%')) 
```

```python
# Cast the string value to float

float(int_rate.strip().strip('%'))
```

```python
# check to see if we have cast string to float

type(float(int_rate.strip('%')))
```

```python
# Let's write a function that can do from above to 
# any string

def int_rate_to_float(cell_contents):
  return float(cell_contents.strip().strip('%'))
```

```python
# Now test the function by calling it on our example

int_rate_to_float(int_rate)
```

```python
# Is our data type correct?

type(int_rate_to_float(int_rate))
```

```python
# Lets apply the solution to every column

new_list = []
for cell in df['int_rate']:
  new_list.append(int_rate_to_float(cell))

df['int_rate_cleaned'] = pd.Series (new_list)
```

```python
# What type of data is in our new column/

df['int_rate_float'] = df['int_rate'].apply(int_rate_to_float)

df.head()
```

```python
# Convert the term column from string to integer.

print(pd.to_numeric(df['term'], errors='coerce'))

```

```python
# Make a column named loan_status_is_great. It should contain the integer 1 
# if loan_status is "Current" or "Fully Paid." Else it 
# should contain the integer 0.

def loan_great():
   if (df['loan_status']).any == 'Current' or (df['loan_status']).any == 'Fully Paid':
     return 1
   else:
     return 0

df['loan_status_is_great']=df['loan_status'].apply(loan_great())

```
